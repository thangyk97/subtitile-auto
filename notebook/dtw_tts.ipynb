```python
import librosa
import matplotlib.pyplot as plt
from dtw import dtw
from numpy.linalg import norm
import numpy as np
#
%matplotlib notebook
```

```python
_path = "/home/thangnd/git/python/subtitile-auto/"
```

```python
with open('../data/lyric1.txt', 'r') as _file:
    l = _file.readlines()
    l = [s.replace('\n', '').replace('\u3000', '') for s in l]
```

```python
_len = 0
y = []
M = {}
for i in range(len(l)):
    _y, _sr = librosa.load(_path + 'data/l1/wav/1-' + str(i) +'.wav')
    y.append(_y)
    M[i] = [_len, _len + _y.shape[0]]
    _len += _y.shape[0]

print("sr of convert = ", _sr)
y = np.concatenate(y, axis=0)
```

```python
r, sr = librosa.load(_path + 'data/l1/wav/1.wav', offset=27, duration=170)
if sr != _sr:
    print("CAUTION: 2 AUDIO DON'T HAVE SAME RATE")
print("sr of real = ", sr)
```

```python
print("Convert length = \t", y.shape[0])
print("Real length = \t\t", r.shape[0])
```

```python
mfcc_r = librosa.feature.mfcc(r, sr)
mfcc_y = librosa.feature.mfcc(y, _sr)
```

```python
dist, cost, acc_cost, path = dtw(mfcc_y.T, mfcc_r.T, dist=lambda x, y: norm(x - y, ord=1))
print('Normalized distance between the two sounds:', dist)
```

```python
plt.figure()
plt.imshow(cost.T, origin='lower', interpolation='nearest')
plt.plot(path[0], path[1], 'w')
plt.xlim((-0.5, cost.shape[0]-0.5))
plt.ylim((-0.5, cost.shape[1]-0.5));
```

```python
path = np.concatenate(path, axis=0)
path = np.reshape(path, (2, -1))
```

```python
M1 = []
for k in M:
    M1.append([M[k][0] // 512, M[k][1] // 512])
```
